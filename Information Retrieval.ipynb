# Removing Punctuation and converting to lower case

```python
punctuation = """!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""
Final_DF['txt_without_punc'] = Final_DF['Summary & Research areas'].str.lower().apply(lambda x: ''.join(ch for ch in x if ch not in set(punctuation)))
Final_DF['txt_without_punc'][0]
```

# Importing Stop words

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words = stopwords.words('english')

print(stop_words)
```

```python
Final_DF['clean_text'] = Final_DF['txt_without_punc'].apply(lambda x: ' '.join([item for item in x.split() if item not in stop_words]))
Final_DF
```

```python
from nltk.stem.snowball import SnowballStemmer

# Use English stemmer.
stemmer = SnowballStemmer("english")

Final_DF['cleartext'] = Final_DF['clean_text'].str.split().apply([lambda x: [stemmer.stem(y) for y in x]]) # Stem every word.
```

```python
Final_DF
```

```python
Final_DF['Stemm_text'] = Final_DF["cleartext"].str.join(" ") 
```

```python
Final_DF
```

```python
import nltk

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(Stemm_text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(Stemm_text)]

Final_DF['text_lemmatized'] = (Final_DF['Stemm_text'].apply(lemmatize_text)).str.join(' ')
```

# Indexing the text

```python
lt_ws = []
for lt in Final_DF['cleartext']:
    lt_ws.append(lt)
    
for lt in lt_ws:
    indexing = {wrd : [(lt_ws.index(lt))+1] for wrd in lt}
    
indexing
```

```python
from collections import defaultdict

class invertedIndex(object):
    def __init__(self,docs):
        self.docSets = defaultdict(set)
        for index, doc in enumerate(docs):
            for term in doc.split():
                self.docSets[term].add(index)
                
        print(self.docSets)
        
    def search(self,term):
        return self.docSets[term]

    
docs=[sent for sent in Final_DF.clean_text]
i=invertedIndex(docs)
```
